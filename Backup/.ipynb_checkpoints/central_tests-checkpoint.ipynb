```python
x = list(combinations_with_replacement(list(range(2,25)), 2))
for i in x:
    print(i)
```

```python
expectation_cut = 0
expectation_bests = pd.DataFrame()
bests_results_lr = pd.DataFrame()
cut_lr = 0.0
for i in x: 
    
    clear_output(wait=True)
    print(f'{i}')
    
    data = my_data
    smothing1 = i[0]
    smothing2 = i[1]
    smothing3 = 8
    data['eur_pmo'], data['eur_signal'] = indi.PMO(data['eur'],smothing1,smothing2,smothing3)
    data['usd_pmo'], data['usd_signal'] = indi.PMO(data['usd'],smothing1,smothing2,smothing3)
    data['gbp_pmo'], data['gbp_signal'] = indi.PMO(data['gbp'],smothing1,smothing2,smothing3)
    data['jpy_pmo'], data['jpy_signal'] = indi.PMO(data['jpy'],smothing1,smothing2,smothing3)
    data['cad_pmo'], data['cad_signal'] = indi.PMO(data['cad'],smothing1,smothing2,smothing3)
    data['aud_pmo'], data['aud_signal'] = indi.PMO(data['aud'],smothing1,smothing2,smothing3)
    data['nzd_pmo'], data['nzd_signal'] = indi.PMO(data['nzd'],smothing1,smothing2,smothing3)
    data['chf_pmo'], data['chf_signal'] = indi.PMO(data['chf'],smothing1,smothing2,smothing3)
    data = dt.pmo_signals(data, signal=False)
    
    y = bt.pct_backtest_without_tpsl(data, risco=0.01, balance=1000)
    
    profit = []
    loss = []
    
    for j in y.iloc[:,3:].sum(axis=1):
        if j < 0:
            loss.append(j)
        else:
            profit.append(j)
            
    profit = pd.Series(profit)
    loss = pd.Series(loss)
    prob_won = len(profit) / len(y)
    prob_loss = len(loss) / len(y)
    expectation = (prob_won * profit.mean()) + (prob_loss * (loss.mean()))
    
    if expectation > expectation_cut:
        expectation_bests[f'{i}'] = y['result']
        expectation_cut = expectation
    
    lr_test = pd.DataFrame()
    lr_test['x'] = pd.Series(range(len(y['result'].dropna())))
    lr_test['y'] = y['result'].dropna()
    x_v = lr_test[['x']]
    y_v = lr_test[['y']]
    model = LinearRegression()
    model.fit(x_v, y_v)
    result = model.score(x_v, y_v)

    if result > cut_lr:
        bests_results_lr[f'{i}'] = y['result']
        cut_lr = result
    
```
